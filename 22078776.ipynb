```python
# https://www.kaggle.com/datasets/pushpakhinglaspure/used-car-price-prediction

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cardekho_data.csv')

fig, axs = plt.subplots(2, 2, figsize=(20, 17), facecolor='#E5E5E5')
fig.suptitle('Comprehensive Analysis of Car Listings on CarDekho', fontsize=25, color='darkred', fontweight='bold')
plt.figtext(0.5, 0.94, 'Gadiraju Dinesh - 22078776', ha='center', fontsize=18, color='darkred', fontweight='bold')

axs[0, 0].set_facecolor('lightgrey')
top_10_cars = df['Car_Name'].value_counts().nlargest(10).sort_values(ascending=True)
top_10_cars.plot(kind='barh', color='darkred', ax=axs[0, 0])
axs[0, 0].set_title('Top 10 Cars Listed on CarDekho Website', fontsize=16, color='darkred', fontweight='bold')
axs[0, 0].set_xlabel('Count', fontsize=13, color='darkred', fontweight='bold')
axs[0, 0].set_ylabel('Car Name', fontsize=13, color='darkred', fontweight='bold')

axs[0, 1].set_facecolor('lightgrey')
sns.histplot(df['Year'], kde=True, color='darkred', fill=True, ax=axs[0, 1])
axs[0, 1].set_title('Manufacturing Year of the Cars Listed on CarDekho Website', fontsize=16, color='darkred', fontweight='bold')
axs[0, 1].set_xlabel('Manufacturing Year', fontsize=13, color='darkred', fontweight='bold')
axs[0, 1].set_ylabel('Density', fontsize=13, color='darkred', fontweight='bold')
plt.subplots_adjust(hspace=0.3)

axs[1, 0].set_facecolor('lightgrey')
axs[1, 0].scatter(df['Selling_Price'], df['Present_Price'], alpha=0.8, c='darkred')
axs[1, 0].set_title('Impact of Present Price on Selling Price', fontsize=16, color='darkred', fontweight='bold')
axs[1, 0].set_xlabel('Selling Price in Lakhs (Rupees)', fontsize=13, color='darkred', fontweight='bold')
axs[1, 0].set_ylabel('Present Price in Lakhs (Rupees)', fontsize=13, color='darkred', fontweight='bold')
axs[1, 0].grid(False)

axs[1, 1].set_facecolor('lightgrey')
fuel_type_counts = df['Fuel_Type'].value_counts().head(5)
explode = (0.1, 0, 0)
colors = ['darkred', 'darkorange', 'black']
axs[1, 1].pie(fuel_type_counts, labels=fuel_type_counts.index, autopct='%1.1f%%', startangle=140,
              explode=explode, colors=colors, wedgeprops=dict(width=0.5))
axs[1, 1].axis('equal')
axs[1, 1].set_title('Fuel Types', fontsize=16, color='darkred', fontweight='bold')
legend_labels = [f'{label} ({percentage:.1f}%)' for label, percentage in zip(fuel_type_counts.index, fuel_type_counts/fuel_type_counts.sum()*100)]
legend = axs[1, 1].legend(legend_labels, title='Fuel_Type', loc='center left', bbox_to_anchor=(0.6, 1))
legend.get_frame().set_facecolor('none')
legend.get_frame().set_linewidth(0)

Analysis = """
Detailed Analysis: The findings highlight Honda City dominance as the most listed car, with Toyota Corolla Altis following closely.

A significant portion of listed cars falls within the 2010-2018 manufacturing year range. 

Notably, there is a linear correlation between listed selling prices and present market prices. 

Petrol cars dominate, constituting 79.4% of listings, while CNG cars have minimal representation at 0.7%. 

These trends suggest a preference for popular models from recent years, indicating market demand alignment. 

The strong presence of petrol cars reflects prevailing fuel preferences, with CNG having a marginal share, possibly due to limited popularity 

or availability in the market."""

fig.text(0.08, -0.12, Analysis, fontsize=15, ha='left', color='darkred', fontweight='bold')

plt.show()
```
